```python
Marchdata= pd.DataFrame(np.random.randint(1,30,120)) #I am using random randint since I need integers from 
#a uniform distribution as the event of making 1 only purchase has the same probability than making 30 purchases in a month for example


Marchdata.reset_index(drop=True, inplace=True)

Marchdata.columns =['March purchases']
Marchdata #This is the purchase frequency we have for the month of March. 

Marchdata.sum()
print(Marchdata.sum())
```
