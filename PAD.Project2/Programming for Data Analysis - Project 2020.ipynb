The original dataset consists of 27 observations on 4 variables which compare intestine length for tadpoles with and without exposure to a fungus called (Batrachochytrium dendrobatidis (Bd)). 

The purpose of this studio was to determine whether tadpoles can adjust the relative length of their intestines if they are exposed to said fungus. 

The 4 Variables that this dataset shows are: 
- Treatment: This is a categorical variable that defines whether the tadpole was exposed to the fungus (Bd=yes) or no (Control=no)

- Body: Represents the length of the animal body in mm. 

- Intestine_Length: Is the length of the intestine in mm.

- MouthpartDamage: Measure of damage to the mouth.


My reason to choose this real world example is that I am really interested on the animal world and I found this was an unusual dataset and have curiosity to see the distribution of it and how we can reproduce it and make assumptions or predictions based on it. 

After having consulted this [source](http://www.int-res.com/articles/dao2014/112/d112p237.pdf), I can see that it seems to be a decrease on the tadpole’s body size when they affected with the fungus as a result of them having difficulties to feed due to mouth damage. 

With my synthetic dataset I hope we can see these relationships reproduced and make conclusions. 

Before I create my synthetic data, let's have a look into the original dataset using 'Pandas’, 'Matplotlib' and 'Seaborn' libraries.


```python
#First, I imported the necessary modules for this section

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading data

tpoles_dataset= pd.read_csv('Tadpoles_dataset.csv',delimiter=',')

tpolesdf=pd.DataFrame(data=tpoles_dataset, columns=['Treatment','Body','Intestine_Length', 'MouthpartDamage'])#using DataFrame 
#with original dataset

print(tpolesdf)
print(tpolesdf.describe())#using this command to find out mean and std for the different variables.
```

## Creating synthetic data from a normal distribution

In this section I am going to create a normal distribution version of this synthetic dataset using 'numpy random normal'.
The 'numpy random normal', module outputs a sample of numbers, in this case I need floats, following a normal disribution. 


```python
#Creating data:

#Using the mean values from our inspiration dataset I am adding these to generate the random samples

np.random.seed(1)# using numpy random seed to maintain same values in samples

Body_measures_withfungus=np.random.normal(17, 1.30, 60)
Body_measures_nofungus=np.random.normal(19.95, 1.99, 60)
Intestine_length=np.random.normal(184.250, 31.80, 120)
Mouth_damage_with_fungus=np.random.normal(0.659, 0.106, 60)
Mouth_damage_nofungus=np.random.normal(0.342, 0.0834, 60)

#Printing these values

print('Random samples for body length (in mm) for tadpoles in contact with the fungus:')
print(Body_measures_withfungus)

print('--------------------------------------------------------')

print('Random samples for body length (in mm) for tadpoles not in contact with the fungus:')
print(Body_measures_nofungus)

print('--------------------------------------------------------')

print('Random samples for intestine length (in mm) for tadpoles:')
print(Intestine_length)

print('--------------------------------------------------------')

print('Random samples for mouth damage for tadpoles in contact with the fungus:')
print(Mouth_damage_with_fungus)

print('--------------------------------------------------------')

print('Random samples for mouth damage for tadpoles not in contact with the fungus:')
print(Mouth_damage_nofungus)
```

### Plotting dataset to observe distribution

As I could observe in the real-world phenomenon, I have created this dataset to follow a normal distribution. 
A normal distribution is often used in these types of events in which we have a list of measurements as in this case, but we cannot look at every measure individually, we do have a range. 
The central point is the mean where the peak of the bell is. 

The normal distribution has some important characteristics such as: 

1. Some value of the uncertain variable is the most likely (the mean of the
distribution).
2. It is symmetrical about the mean.
3. The uncertain variable is more likely to be in the vicinity of the mean than further
away. [source](https://onlinelibrary.wiley.com/doi/pdf/10.1002/9781119197096.app03)


In this case since I have added two different means to the same data, the peak might not be perfect as we have a mean for cases of contact with fungus and different one for no contact cases.
